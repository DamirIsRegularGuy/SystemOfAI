```python
import numpy as np
import matplotlib.pyplot as plt

num_elements = float(input("Введите количество элементов для добавления: "))

matrix = []
for i in range(int(num_elements)*3):
    random_number = np.random.randint(1, 15)  
    symbol = "нолик" if i % 2 == 0 else "крестик"
    matrix.append([random_number, symbol])

matrix_np = np.array(matrix, dtype=object)

print(matrix_np)


evklid_arr = []
for i in range(int(num_elements)*3):
    evklid_arr.append([float(np.sqrt(pow(matrix_np[i][0] - float(num_elements), 2))), i])

evklid_arr_np = np.array(evklid_arr, dtype=object)

matrix_np = evklid_arr_np[np.argsort(evklid_arr_np[:, 0])]

print(evklid_arr_np)
print("\n")
print(matrix_np)

chet = 0
nechet = 0
for i in range(int(num_elements)):
    if(matrix_np[i][1] % 2 == 0): chet += 1
    else: nechet +=1

if(chet > nechet): 
    print("нолик c " + str(chet/100) + "%, чем крестик с " + str(nechet/100) + "%")
else:
    print("крестик c " + str(nechet/100) + "%, чем нолик с " + str(chet/100) + "%")





```
